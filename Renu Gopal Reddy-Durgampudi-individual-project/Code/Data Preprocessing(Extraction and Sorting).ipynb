```python
#count number of files in given directory
import os

def count_files(in_directory):
    joiner= (in_directory + os.path.sep).__add__
    return sum(
        os.path.isfile(filename)
        for filename
        in map(joiner, os.listdir(in_directory))
    )
c1=count_files("C:/Users/renug/Desktop/rvlcdip/validation/1/")
print(c1)
```

```python
#check properties of images in given directory
from os import listdir
from os.path import isfile, join
import numpy
import cv2

mypath='C:/Users/renug/Desktop/rvlcdip/validation/1/'
onlyfiles = [ f for f in listdir(mypath) if isfile(join(mypath,f)) ]
images = numpy.empty(len(onlyfiles), dtype=object)
for n in range(0, len(onlyfiles)):
    images[n] = cv2.imread( join(mypath,onlyfiles[n]) )
    height, width, channels = images[n].shape
    print(height, width, channels)
```
